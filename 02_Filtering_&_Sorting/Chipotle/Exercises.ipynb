### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

### Step 4. How many products cost more than $10.00?

```python
# Set the maximum width of a column to unlimited (None)
# pd.set_option('display.max_colwidth', None)

chipo["unit_price"] = chipo["item_price"].str.strip("$").astype("float64") / chipo["quantity"]
df = chipo.drop_duplicates(["item_name", "choice_description"])
df[df.unit_price > 10].shape[0]
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
df[df.unit_price > 10][["item_name", "choice_description", "unit_price"]]
```

### Step 6. Sort by the name of the item

```python
chipo.drop_duplicates(["item_name", "choice_description"])[["item_name", "choice_description", "unit_price"]].sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.loc[chipo["item_price"].idxmax()]["quantity"]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo.item_name == "Veggie Salad Bowl"].quantity.sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)].order_id.count()
```
